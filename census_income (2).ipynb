**Encoding The categorical variables**

```python
def object_to_int(dataframe):
 for i in dataframe.columns:
  if dataframe[i].dtype=='object':
   le=LabelEncoder()
   dataframe[i]=le.fit_transform(dataframe[i])
 return dataframe

```

```python
df2=df
df2=object_to_int(df2)
```

```python
df2.head()
```
